## Problema 1

#### i) 
Si una estrella tiene una magnitud aparente de +6 y sabemos que se encuentra a una distancia de $8 \ \text{pc}$, ¿cuál es su magnitud absoluta?

**Solución**:

Para comparar las magnitudes de dos estrellas tenemos la ecuación

\begin{equation}
m_1 - m_2 = -2.5 \log_{10} \Big( \frac{F_1}{F_2} \Big)
\end{equation}

donde $m_1$ y $F_1$ son la magnitud aparente y el flujo de una estrella que llamaremos "1"; y $m_2$ y $F_2$ son la magnitud aparente y el flujo de una estrella que llamaremos "2". De manera que la ecuación (1), de cierto modo, compara las magnitudes de las estrellas 1 y 2.

Recordemos también que la luminosidad $L$ y el flujo $F$ de una estrella se relacionan mediante:

$$
L = F \times \text{Área}
$$

De manera que el Flujo puede ser escrito como:

\begin{equation}
F = \frac{L}{\text{Área}} = 4 \pi d^2
\end{equation}

Donde para una esfera de radio $d$ (que es el área que ilumina una estrella alrededor de ella) el área es $4 \pi d^2$.

Reemplazando (2) en (1) tenemos que esto da:

$$
m_1 - m_2 = -2.5 \log_{10} \Big( \frac{\frac{L_1}{4 \pi (d_1)^2}}{\frac{L_2}{4 \pi (d_2)^2}} \Big) = 
-2.5 \log_{10} \Big( \frac{L_1}{L_2} \times \Big[\frac{d_2}{d_1} \Big]^2\Big)
$$

Por lo que llegamos a:

\begin{equation}
m_1 - m_2 = -2.5 \log_{10} \Big( \frac{L_1}{L_2} \times  \Big[ \frac{d_2}{d_1}  \Big]^2\Big)
\end{equation}

Ahora bien, en nuestro caso $d_1$ y $d_2$ son las distancias a las cuales están las estrellas 1 y 2, respectivamente. ¿Por qué? Porque $d_1$ es el radio de la esfera que está iluminando la estrella 1 y $d_2$ es el radio de la esfera que está iluminando la estrella 2 y ambas decaen con el inverso de la distancia al cuadrado (ver imagen de abajo para tener referencia)

La definición de magnitud absoluta, la cual se escribe generalmente con una M mayúscula (y la magnitud aparente con una M minúscula) es: la magnitud absoluta es la magnitud que tiene un objeto al encontrarse a $10 \text {pc}$ de distancia.

Recordemos que en ejercicio queremos encontrar una fórmula para la magnitud absoluta.

Sea $m$ la magnitud aparente de una estrella cuando está a una distancia $d$, y $M$ la magnitud absoluta de esa misma estrella cuando está a $10 \text {pc}$ de distancia; entonces tendremos aplicando la fórmula (3):

$$
m - M = -2.5 \log_{10} \Big( \frac{L}{L} \times \Big[\frac{d_2}{d_1}\Big]^2\Big)
$$

En el paso anterior L se simplifica pues estamos comparando la luminosidad de la estrella consigo misma, así y aplicando propiedades de los logaritmos:

$$
m - M = 5 \log_{10} \Big( \frac{d_1}{d_2} \Big)
$$

Como $d_1$ es una distancia $d$ que desconocemos, pero $d_2$ es la distancia a la cual se encuentra la estrella cuando tiene magnitud absoluta entonces $d_2 = 10 \ \text{pc}$. Así la ecuación de arriba queda:

\begin{equation}
m - M = 5 \log_{10} \Big( \frac{d}{10 \ \text{pc}} \Big)
\end{equation}

Que es equivalente a:

$$
m-M = 5 \cdot \log_{10} \Big( \frac{d}{10 \ \text{pc}} \Big) = 5 \Big[ \log_{10} \Big( \frac{d}{\text{pc}} \Big) - \log_{10}(10) \Big]= 5 \Big[\log_{10} \Big( \frac{d}{\text{pc}} \Big) - 1\Big]  = 5 \cdot \log_{10} \frac{d}{\text{pc}} - 5
$$

Que finalmente da:

$$
\frac{m-M+5}{5} = \log_{10}\Big( \frac{d}{\text{pc}} \Big)
$$

Elevando todo a $10$ para despejar lo que está dentro del paréntesis llegamos finalmente a que la distancia es:

\begin{equation}
d = 10^{\frac{m-M+5}{5}} \ \text{pc}
\end{equation}

Ahora que  tenemos todas las ecuaciones que nos sirven podemos responder a lo que nos preguntan: "Si una estrella tiene una magnitud aparente de +6 y sabemos que se encuentra a una distancia de $8 \ \text{pc}$, ¿cuál es su magnitud absoluta?"

Para ello simplemente podemos mirar la ecuación $(4)$ y darnos cuenta de que podemos despejar la magnitud absoluta $M$ simplemente como:

$$
M = m - 5 \log_{10} \Big( \frac{d}{10 \ \text{pc}} \Big)
$$

Lo cual nos da:

```python
import numpy as np
from astropy import units as u


def absolute_magnitude(apparent_magnitude, distance):
    """
    Returns absolute magnitude given an apparent magnitude and a distance (in parsecs)
    """
    return apparent_magnitude - 5*np.log10((distance.to(u.pc)) / (10*u.pc))


### Input values
distance_star = 8 * u.pc
app_mag_star = 6

### Results
absolute_mag = absolute_magnitude(app_mag_star, distance_star)

print(f'Una estrella con magnitud aparente {app_mag_star:.2f} y a una distancia de {distance_star:.2f}', end = ' ')
print(f'tendrá una magnitud absoluta de {absolute_mag:.2f}')
```

#### ii)
Hay ciertos objetos en el universo que nos permiten medir distancias en el universo. Entre estos objetos están las Supernovas tipo Ia. Nos sirven porque gracias a investigaciones sabemos qué magnitud absoluta tendrá -aproximadamente- a $10 \ \text{pc}$ de distancia, la cual será de más menos $M_\text{v} = -19.3$. Esto hace que en astronomìa sean conocidas como candelas estándar (objetos que sirven para estimar distancias en astrofísica); ya que, si conocemos previamente la magnitud absoluta de un objeto y logramos medir su magnitud aparente, entonces tenemos la distancia. Si ocurre una supernova cuya magnitud aparente es $m_\text{v} \sim 6$, ¿a qué distancia se encuentra la supernova?

**Solución**:

Para esto simplemente utilizamos la ecuación $(5)$ hallada anteriormente. Esto nos dará que la Supernova está a una distancia de:

```python
def approximated_distance(abs_magnitude, app_magnitude):
    """
    Returns an approximated sitance to an object (ignores reddening)
    given its apparent and absolute magnitudes
    """
    exponent = (app_magnitude - abs_magnitude + 5)/5.
    return np.power(10, exponent) * u.pc

### Input
m_mag = 8
M_mag = -19.3 # absolute mag for Supernovas Ia

### Results
approx_dist = approximated_distance(M_mag, m_mag)

print(f'Para un objeto con magnitud aparente {m_mag:.2f} y magnitud absoluta {M_mag:.2f}, éste se encuentra', end = ' ')
print(f'a una distancia de {approx_dist:.2e}')

```

### f)
El ojo humano puede ver hasta una magnitud de $m_\text{v} \sim 5$. ¿Cuál es la máxima distancia a la que podría estar el Sol/una estrella como el Sol para que pudiésemos verlo a ojo desnudo si el Sol tiene una magnitud aparente de $m_{\odot ,\text{v}} \sim -26.74$? Hint: Recuerde que primero debe conocer la magnitud absoluta del Sol y que una unidad astronómica ($1 \ \text{AU}$) equivale a $4.84 \times 10^{-6} \ \text{pc}$ (pársecs)...

**Solución**:

Ya que nos piden saber qué tan lejos podría estar el Sol para verlo a ojo desnudo (sin telescopio) debemos hacer lo siguiente:

- Calcular la magnitud absoluta del Sol sabiendo que el Sol está a 1 unidad astronómica de distancia (no olvide convertir unidades de unidad astronómica a pársec) usando
$$
M = m - 5 \log_{10} \Big( \frac{d}{10 \ \text{pc}} \Big)
$$

- Una vez obtenida la magnitud absoluta del Sol, ver a qué distancia estaría el Sol si tuviese una magnitud aparente de $m_\text{V} \sim 5$. Si el Sol/una estrella como el Sol estuviese más lejos que esa distancia, no la podríamos ver a ojo desnudo; si estuviese más cerca entonces podríamos seguir viéndola sin problemas.

```python
from astropy import constants as const


###Input

app_mag_Sun = -26.74
lowest_app_mag = 5 # minimum apparent magnitud human eye can detect

### Results
abs_mag_Sun = absolute_magnitude(app_mag_Sun, const.au)
max_distance_possible = approximated_distance(abs_mag_Sun, lowest_app_mag)

print(f'El Sol/una estrella como el Sol tendrá una magnitud absoluta de {abs_mag_Sun:.2f}')
print(f'Por lo que lo máximo que podría estar el Sol de lejos es {max_distance_possible:.2e}')
print('Si estuviera más lejos que eso no lo veríamos; si estuviera más cerca de eso lo seguiríamos viendo')
```
